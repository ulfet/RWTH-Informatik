# Temporal networks

Consider a human contact dataset (highschool_2011.csv) which consists of contacts between highschool students. Column 1 indicates the time and columns 2 and 3 indicates the node ids (you can ignore the rest of the columns). 

Write a function that takes as input the dataset and calculates the temporal correlation present in the dataset. 

```python
import pandas as pd
data = pd.read_csv("highschool_2011.csv", sep="\t", names=["time", "id1", "id2", "x", "y"])
data
```

```python
# generating all contacts for each time stamp
def groupFunction(dataset):
    timeDict = {}
    for index, row in dataset.iterrows():
        t = row["time"]
        timeDict[t] = []
    
    for index, row in dataset.iterrows():
        t = row["time"]
        aList = set([row["id1"], row["id2"]])
        timeDict[t].append(aList)
    return timeDict

groupedOnTimeDict = groupFunction(data)
print(groupedOnTimeDict)
```

```python
from math import sqrt
def getMeContactsFor(xid, timeContactList):
    resultList = set()
    for aSet in timeContactList:
        if xid in aSet:
            resultList.add(frozenset(list(aSet)))
    return resultList

anna = 0
def countMatches(set1, set2):
    global anna
    if anna > 0:
        print(len(set1  & set2))
        anna -= 1
    return len(set1  & set2)

def findCorrelation(timeDict):
    # to use for indexing
    allKeys = list(timeDict.keys())
    allKeysSorted = sorted(allKeys)
    
    minTime = allKeysSorted[0]
    maxTime = allKeysSorted[-1]
    
    listOfAllPossibleTimes = list(range(minTime, maxTime+20, 20))
    lenOfAllPossibleTimes = len(listOfAllPossibleTimes)
    
    allIDs = range(1,127)
    lenAllIDs = len(list(allIDs))
    
    outerSum = 0.0
    for xid in allIDs:
        innerSum = 0.0
        for (index, time) in enumerate(allKeysSorted):
            ct1 = getMeContactsFor(xid, timeDict[time])
            time2 = time+20
            if time2 not in allKeysSorted:
                continue
            ct2 = getMeContactsFor(xid, timeDict[time2])
            if len(ct1) != 0 and len(ct2) != 0:
                matchCount = countMatches(ct1, ct2)
                innerSum += matchCount / sqrt(len(ct1)*len(ct2))
                
        innerSum /= (lenOfAllPossibleTimes - 1)
        outerSum += innerSum
        
    outerSum /= lenAllIDs
    return outerSum
        
correlationValue = findCorrelation(groupedOnTimeDict)
print(correlationValue)
```

```python
def calculate_correlation(dataset):
    timeDict = groupFunction(dataset)
    result = findCorrelation(timeDict)
    return result
```

```python
import random
# which form does this network have?
# EXCPECTS: pandas dataframe as in highschool 2011
def shuffle_network(network):
    network = network[["time", "id1", "id2"]]
    shuffled_network = network.copy()
    edges = shuffled_network.shape[0]
    for step in range(0, 1000):
        edge1 = shuffled_network.sample()
        edge2 = shuffled_network.sample()
        tmp = edge1["time"]
        shuffled_network.loc[edge1.index[0]][0] = edge2["time"]
        shuffled_network.loc[edge2.index[0]][0] = tmp
    return shuffled_network
```

From the contact information provided in the dataset write a function to calculate the activity potential of each node i (F(i)). Now write a function to generate the network for next time step using the activity-driven network model. You can set the value of m (the number of links generated by each active node to 2). Note that a node i becomes active with a probability alpha\*F(i), alpha = 10

```python
def activity_potential(network):
    potential = {}
    timeDict = groupFunction(network)
    for idx in range(1,127):
        #node_potential = 0
        num_pot = 0
        den_pot = 0
        for time_step in list(timeDict.keys()):
            # set(set(), set())
            contacts_at_t = timeDict[time_step]
            for contact in contacts_at_t:
                if idx in contact:
                    num_pot+=1
            den_pot+=len(contacts_at_t)
        potential[idx] = num_pot/den_pot
    return potential
```

Write a function to obtain the ego (immediate neighbors) for each node. Your function should take as input the network and the node id and return its ego.
Using the activity potential calculated previously write a function to generate the network for the next time step using the activity-driven network model with memory. Any node links with a previously contacted node with probability n/n+1 (n is the size of its ego) or with a new node with probability 1/n+1. Note that a node i becomes active with a probability alpha\*F(i), alpha = 10

```python
def getEgo(networkVar, nodeId):
    timeDict = groupFunction(networkVar)
    neighborSet = set()
    for time in timeDict.keys():
        for contact in timeDict[time]:
#             print(contact)
            if nodeId in contact:
                newID = contact.difference(set([nodeId]))
#                 print(contact, newID)
                neighborSet = neighborSet.union(newID)
    return neighborSet

def activity_step_memory(network, alpha):
    potential = activity_potential(network)
    timeDict = groupFunction(network)
    max_time = max(list(timeDict.keys()))
    contacts = set()
    entireSet = set(list(range(1,127)))
    
    for idx in range(1,127):
        node_active = False
        # random function to select node_active with prob alpha*potential
        prob = alpha*potential[idx]
        # print(prob)
        node_active = np.random.choice([True,False],p=[prob, 1-prob])
        if node_active:
            print(idx, "was activated")
            egoSet = getEgo(network, idx)
            n = len(egoSet)
            prob2 = n/(n+1)
            shouldIChooseFromEgoSet = np.random.choice([True,False],p=[prob2, 1-prob2])
            for m in range(0,1):
                if shouldIChooseFromEgoSet:
                    idy = random.sample(egoSet, 1)[0]
                else:
                    print("Wow: I choose a new node")
                    remainingSet = entireSet - egoSet
                    idy = random.sample(remainingSet, 1)[0]
                
                contact = frozenset([idx,idy])
                contacts.add(contact)
    return contacts
```
